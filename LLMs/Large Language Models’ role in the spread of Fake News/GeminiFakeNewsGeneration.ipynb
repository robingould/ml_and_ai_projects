```python
import os, csv, json, time
from datetime import datetime
import pandas as pd
import numpy as np
import google.generativeai as genai
import logging

# Configure logging

# get api and required data

with open("googleapi", 'r') as f:
    gemini_key = f.read().strip()
'''
df = pd.read_csv("data/test6000.csv")
texts = df['text'].to_numpy()
labels = df['label'].to_numpy()
'''
'''with open("code/text.txt", 'r') as f:
    text = f.read().strip()'''

# set up the gemini prompt stuff
genai.configure(api_key = gemini_key)
model = genai.GenerativeModel('gemini-pro')

prompt = 'Negate this text, completely change it so that it has the opposite meaning and rewrite it in a formal tone:'
safe = [
    {
        "category": "HARM_CATEGORY_HARASSMENT",
        "threshold": "BLOCK_NONE",
    },
    {
        "category": "HARM_CATEGORY_HATE_SPEECH",
        "threshold": "BLOCK_NONE",
    },
    {
        "category": "HARM_CATEGORY_SEXUALLY_EXPLICIT",
        "threshold": "BLOCK_NONE",
    },
    {
        "category": "HARM_CATEGORY_DANGEROUS_CONTENT",
        "threshold": "BLOCK_NONE",
    },
]



with open("texts", 'r') as f:
  text = f.read().strip()

print(f"Original text:\n{text}\n")
try:
  response = model.generate_content(f"{prompt} {text}", safety_settings=safe)
  print(f"altered text:\n{response.text.strip()}")

except Exception as e:
  print(f"failed {e}")
```
